```python
df['type'].value_counts()
```

```python
# -------------------BAR CHART---------------------
import matplotlib.pyplot as plt
a=df['type'].value_counts()
plt.title('No Of Movies & No Of TV Shows')
plt.xlabel('Movie & TV Show')
plt.ylabel('No Of Shows')
plt.bar(a.index,a.values)
plt.show()
```

```python
x=df['rating'].value_counts().head(5)
x
```

```python
#-----------PIE CHART------------------------

import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))

x=df['rating'].value_counts().head(5)
plt.title('Movie Ratings')
plt.pie(x.values,labels=x.index,autopct='%1.2f%%',explode=[0,0,0.2,0,0],shadow=True)
plt.show()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%',explode=[0,0.2],shadow=True)
plt.figure(figsize=(10,8))

plt.show()
```

```python
#--------------------BAR PLOT--------------------------
import matplotlib.pyplot as plt
x=df['country'].value_counts().head(10)
plt.figure(figsize=(12,5))
plt.bar(x.index,x.values)
plt.xlabel('Countrys')
plt.ylabel('No Of Shows release')
plt.title('Top 10 Countrys')
plt.show()
```

```python
#-------------HISTOGRAM-----------------------

plt.hist(df['release_year'], bins=20)

plt.title('Release Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# visualisation
a=df['release_year'].value_counts().sort_index()
a
```

```python
#---------------------LINE FLOT--------------------

plt.figure(figsize=(14,5))
plt.plot(a.index,a.values,marker='o')
plt.title('Release Year',color='red')
plt.xlabel('Year')
plt.ylabel('No Of Shows release')
plt.show()
```

```python
#-----------Top 10 Directors--------------------


df['director'].value_counts().head(10).plot(kind='barh')

plt.title('Top Directors')
plt.show()
```
